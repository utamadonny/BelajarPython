#### Model Prediksi
original sigmoid
$$sig(t)=\frac{1}{1+e^{-t}}$$
Modifikasi
$$sig(t)=\frac{c}{1+e^{-(t-t_0)/a}}$$

```python
puncak = 8
x=np.linspace(-8,8,100)
y=puncak/(1+np.exp(-x))
plt.plot(x,y,label="Sigmoid Function")
plt.grid()
plt.show()
```

```python
def sigmoid_curve(t,a,t0,c):
	"""Modified Sigmoid Function

	Args:
		t (int): Jumlah Hari
		a (int):  Koefisien Curve Fitting
		t0 (int): Jumlah Kasus Awal
		c (int): Nilai Puncak

	Returns:
		float: Nilai Sigmoid Curve
	"""
	return c/(1+np.exp(-(t-t0)/(a)))
```

```python
x=list(df_indonesia.index)
y=list(df_indonesia["total_cases"])
```

```python
from scipy.optimize import curve_fit, fsolve
```

```python
fit = curve_fit(sigmoid_curve,x,y,method="trf")
varA, varB = fit
```

```python
varA
```

```python
varB
```

```python
std_err = np.zeros(len(varA))
for i in range(len(varA)):
	std_err[i] = np.sqrt(varB[i][i])
```

```python
std_err
```

```python
a=varA[0]+std_err[0]
t0=varA[1]+std_err[1]
c=varA[2]+std_err[2]
```

```python
def puncak(x):
	return sigmoid_curve(x,a,t0,c)-int(c)
```

```python
n_puncak=int(fsolve(puncak,t0))
```

```python
print("Hari puncak wabah covid-19 adalah hari ke-" + str(n_puncak))
```

#### Analisis

```python
n_0=max(x)+1
pred_x=list(range(n_0,n_puncak))
```

```python
print("Kita memiliki " + str(len(pred_x)) + " hari untuk mencapai puncak wabah covid-19")
```

```python
pred_y=np.zeros(len(x+pred_x))
for i in range(n_puncak):
	pred_y[i]=sigmoid_curve(i,a,t0,c)
```

```python
print("Prediksi Jumlah Orang pada puncak wabah covid-19 adalah " + str(int(pred_y[n_puncak-1])))
```

```python
plt.plot(x+pred_x,pred_y,label="Prediksi")
plt.scatter(x,y,label="Data Asli", s=10, c="red")
plt.xlabel("Hari")
plt.ylabel("Jumlah Kasus")
plt.legend(loc="best")
plt.grid()
plt.show()
```

### R2 Skor

$$ R^2 = \frac{S_res}{S_tot} $$
$$ R^2 = 1 - \frac{\sum{(y_i-\hat{y_i})^2}}{\sum{(y_i-\hat{y})^2}} $$

```python
def akurasi(y_true,y_pred,x):
	atas=sum((y_true-y_pred[0:len(x)])**2)
	bawah = sum((y_true-np.mean(y))**2)
	r=1-(atas/bawah)
	return r
```

```python
akur=akurasi(y,pred_y,x)
```
